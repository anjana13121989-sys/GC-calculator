```python
#takes dna string and calculates its %gc content
def gc_content(sequence):
    g_count = 0
    c_count = 0
    for base in sequence:
        if base == 'G':
            g_count = g_count + 1
        elif base == 'C':
            c_count = c_count + 1
            
    total_gc = c_count + g_count
    per_content = (total_gc)/len(sequence)*100

    return per_content
    
gc = gc_content('sequence')
print("The %GC content in this:",round(gc,3))
```
