```python
# takes dna string and calculates its properties
def seq_prop(sequence):
    g_count = 0
    c_count = 0
    a_count = 0
    t_count = 0
    for i in sequence:
        if i == 'G':
            g_count = g_count + 1
        elif i == 'C':
            c_count = c_count + 1
        elif i == 'A':
            a_count = a_count + 1
        elif i == 'T':
            t_count = t_count + 1
            
    total_at = a_count + t_count
    total_gc = c_count + g_count
    melt_temp = 4* total_gc + 2 * total_at
    gc_content = (total_gc)/len(sequence)*100
    aneal_temp = melt_temp - 5 
    

    return melt_temp, gc_content, aneal_temp
    
tm,gc,ta = seq_prop("sequence")
print("The melting temperature of sequence in this(째C):",round(tm,3))
print("the gc% of the sequence is(%):",round(gc,3))
print("the anealing temperature is(째C):",round(ta,3))
```
